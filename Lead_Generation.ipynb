## Dataset Overview

The dataset contains the following attributes:

- **Lead_ID**: A unique identifier for each student.

- **Year of Study**: The current year of study of the student (e.g., 1st year, 2nd year, etc.).
- **College Type**: Indicates whether the college is located in a **rural** or **urban** area.

Each record corresponds to a unique student, providing valuable insights into lead demographics and sourcing efficiency.

To allocate the marketing budget, a fixed value of **2 units** per lead was assigned. This decision ensures that every lead, irrespective of its source, location, or program interest, receives an equal share of the budget. The total budget for the marketing campaign was then calculated by multiplying the number of leads by the allocated value of 2. 

This approach simplifies the budget allocation process, enabling efficient resource distribution while allowing for further analysis of lead sources and program interests without complicating the financial aspect of the project. üéØ

By using a consistent budget per lead, the project can focus on evaluating the performance of different lead sources and program interests without the need for complex financial calculations. üîç
